## Quel est le contexte du jeu de données

Le site Twitch est une plateforme de streaming en direct où des créateurs de contenu partagent leurs activités, souvent en jouant à des jeux vidéo, tout en interagissant avec leur audience via des chats en direct.

Notre jeu de données représente les utilisateurs et les streamers actifs sur Twitch. Chaque enregistrement correspond à un instantané capturé toutes les 10 minutes sur une période de 43 jours, incluant des informations telles que les streamers actifs, les utilisateurs présents dans leurs chats respectifs, et leurs interactions.

Ces données sont collectées dans le but de mieux comprendre les dynamiques d'engagement des utilisateurs, les habitudes des spectateurs, et les tendances générales sur la plateforme Twitch.

```python
### Chargement du dataset

import pandas as pd

data = pd.read_csv("Twitch.csv")
data.columns = ['user_id', 'stream_id', 'streamer_username',
'start', 'stop']

data.head(10)
```

```python
data.columns

# Affiche le type des colonnes et le nombre de valeurs non nulles
data.info()
```

```python
### Nettoyage des données
data = data.dropna()
data = data.drop_duplicates()
data = data.iloc[:-3000000]

data.info()

### télécharger data cleaned
data.to_csv("Twitch_clean.csv", index=False)
```
